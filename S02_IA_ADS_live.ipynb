```python
# Primeiro modelo de aprendizado supervisionado
# Regressão linear
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Coleta de dados
df = pd.read_csv("alunos.csv")

# Variáveis independentes e dependentes
X = df[["horas_estudo"]]
y = df["nota"]

# Treinamento do modelo de Regressão Linear
modelo = LinearRegression()
modelo.fit(X, y)

# Coeficientes
print(f'Inclinação: {modelo.coef_[0]}')
print(f'Intercepto: {modelo.intercept_}')

# Gráficos
y_pred = modelo.predict(X)
plt.scatter(df["horas_estudo"], df["nota"], label="Dados reais")
plt.plot(df["horas_estudo"], y_pred, label="Regressão linear", color="red")
plt.xlabel("Horas de estudo")
plt.ylabel("Nota")
plt.legend()
plt.show()


# Previsão
horas = [[6]]
nota_prevista = modelo.predict(horas)
print(f'Para um estudo de {horas} horas, a previsão da nota é de {nota_prevista}.')
```

```python
# Primeiro modelo de aprendizado supervisionado
# Regressão linear
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Coleta de dados
df = pd.read_csv("alunos.csv")

# Variáveis independentes e dependentes
X = df[["faltas"]]
y = df["nota"]

# Treinamento do modelo de Regressão Linear
modelo = LinearRegression()
modelo.fit(X, y)

# Coeficientes
print(f'Inclinação: {modelo.coef_[0]}')
print(f'Intercepto: {modelo.intercept_}')

# Gráficos
y_pred = modelo.predict(X)
plt.scatter(df["faltas"], df["nota"], label="Dados reais")
plt.plot(df["faltas"], y_pred, label="Regressão linear", color="red")
plt.xlabel("Faltas")
plt.ylabel("Nota")
plt.legend()
plt.show()


# Previsão
faltas = [[6]]
nota_prevista = modelo.predict(faltas)
print(f'Para um total de {faltas} faltas, a previsão da nota é de {nota_prevista}.')
```
